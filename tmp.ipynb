```python
import json
import os
import cv2
import matplotlib.pyplot as plt
import random
from tqdm import tqdm
```

```python
json_file = json.load(open('./datasets/error_det/label_train.json'))
images = os.listdir('./datasets/error_det/images-origin')
```

```python
for num,item in tqdm(enumerate(json_file),total = len(images)):
    img_name = item['id']
    img_box = item['region']
    img_path = os.path.join('./datasets/error_det/images-origin',img_name)
    img = cv2.imread(img_path)
    h,w = img.shape[:2]
    cv2.imwrite(f'datasets/error_det/images/train/{num:06d}.png',img)
    with open(f'datasets/error_det/labels/train/{num:06d}.txt','w') as f:
        text = []
        for box in img_box:
            x1, y1, x2, y2 = box
            x1 = x1/w
            y1 = y1/h
            x2 = x2/w
            y2 = y2/h
            w,h = abs(x1-x2),abs(y1-y2)
            x = x1+w/2
            y = y1+h/2
            text.append(' '.join(map(str,[0,x,y,w,h])))
        f.writelines('\n'.join(text))
```

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "7"
from ultralytics import YOLO
# Create a new YOLO model from scratch
model = YOLO("./config/yolov8x.yaml")
# Load a pretrained YOLO model (recommended for training)
model = YOLO("yolov8x.pt")
# Train the model using the 'coco8.yaml' dataset for 3 epochs
results = model.train(data="datasets/error_det/error_det.yaml", epochs=30)
# Evaluate the model's performance on the validation set
# results = model.val()
# Perform object detection on an image using the model
# results = model("https://ultralytics.com/images/bus.jpg")
```
